### Подключаем библиотеки и загружаем данные

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('HWS.csv')
```

```python
print(data.head())
print(data.shape)
```

### Гистограмма

```python
data['Height'].hist()
plt.show()
```

```python
data['Height'].hist(bins=12)
plt.show()
```

### Столбчатая диаграмма

```python
data_gr_height = data.groupby('Index')['Height'].mean()
print(data_gr_height)
```

```python
data_gr_height.plot(kind='bar', title='Средний рост по BMI')
plt.show()
```
