### Подключаем библиотеки и загружаем данные

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('HWS.csv')
```

```python
print(data.head())
print(data.shape)
```

### Гистограмма

```python
data['Height'].hist()
plt.show()
```

```python
data['Height'].hist(bins=12)
plt.show()
```

### Столбчатая диаграмма

```python
data_gr_height = data.groupby('Index')['Height'].mean()
print(data_gr_height)
```

```python
data_gr_height.plot(kind='bar', title='Средний рост по BMI')
plt.show()
```

### Загрузка датасета AQ

```python
aq = pd.read_csv('aq.csv')
print(aq)
```

### Линейный график

```python
aq.plot(x = 'Date', y = 'NO2', rot=20)
plt.show()
```

### Диаграмма рассеяния

```python
data.plot(x='Height', y='Weight', kind='scatter')
plt.show()
```

### Добавление уровней

```python
data[data['Gender'] == 'Male']['Height'].hist(bins=25, alpha = 0.6)
data[data['Gender'] == 'Female']['Height'].hist(bins=25, alpha = 0.6)
plt.title('Гистограмма распределения по росту от пола')
plt.legend(['Male', 'Female'])
plt.show()
```
